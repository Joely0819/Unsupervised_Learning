```python
# Initial imports
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

# Data Preparation

```python
# Data loading
file_path = Path("../Resources/crypto_data.csv")
df_crypto = pd.read_csv(file_path)
df_crypto.head(10)
```

```python
#List the DataFrame's data types to ensure they're aligned to the type of data stored on each column.
df_crypto.dtypes
 
```

```python
#Discard all cryptocurrencies that are not being traded. In other words, filter for currencies that are currently being traded. Once you have done this, drop the IsTrading column from the dataframe.
# Keep only cryptocurrencies that are trading
df_crypto = df_crypto.loc[df_crypto['IsTrading'] == True]
# Keep only cryptocurrencies with a working algorithm
df_crypto = df_crypto.loc[df_crypto['Algorithm'] != 'N/A']
# Remove the 'IsTrading' column
df_crypto = df_crypto.drop(columns = 'IsTrading')
```

```python
#Remove all rows that have at least one null value.
df_crypto = df_crypto.dropna()
df_crypto.isnull().sum()
```

```python
#Filter for cryptocurrencies that have been mined. That is, the total coins mined should be greater than zero.
df_crypto = df_crypto.loc[df_crypto['TotalCoinsMined'] != 0]

# Drop rows where there are 'N/A' text values
df_crypto = df_crypto[df_crypto!='N/A']


```

```python
#'CoinName'column DataFrame  
coinname_df = pd.DataFrame(
    data = df_crypto, columns = ['CoinName'])
coinname_df
```

```python
#dataset to be comprehensible to a machine learning algorithm, its data should be numeric. Since the coin names do not contribute to the analysis of the data, delete the CoinName from the original dataframe.

df_crypto = df_crypto.drop(columns = 'CoinName')
df_crypto.head()
```

```python
#convert the remaining features with text values, Algorithm and ProofType, into numerical data
# Create dummy variables for text features
dummy_crypto= pd.get_dummies(df_crypto, columns=['Algorithm', 'ProofType'])
dummy_crypto.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(dummy_crypto[['TotalCoinSupply', 'TotalCoinsMined']])
print (scaled_data[0:5])
```

```python
scaled_data = scaled_data[~np.isnan(scaled_data).any(axis=1)]
np.isnan(scaled_data).sum()
```

# Dimensionality Reduction

```python
#Perform dimensionality reduction with PCA; reduce to 2 components
from sklearn.decomposition import PCA

reduced_pca = PCA(n_components=2)

crypto_pca = reduced_pca.fit_transform(scaled_data)
```

```python
# Create a DataFrame
pca_df = pd.DataFrame(
    data=crypto_pca, columns=["PC 1", "PC 2"]
)
pca_df.head()
```

```python
from sklearn.cluster import KMeans

# Finding the best value for k
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pca_df)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
 # Predicting clusters with k=2

# Initialize the k-means model
model = KMeans(n_clusters=2, random_state=0)

# Fit the model
model.fit(pca_df)

# Predict clusters
predictions = model.predict(pca_df)

# Add the predicted class column to the dataframe
pca_df["class"] = model.labels_
pca_df.head()
```

```python
# Visualize the clusters
plt.scatter(x=pca_df['PC 1'], y=pca_df['PC 2'], c=pca_df['class'])
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')
plt.title('Clusters')
plt.show()
```
